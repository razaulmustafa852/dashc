```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__) # requires version >= 1.9.0
import cufflinks as cf
init_notebook_mode(connected=True)
# For offline use
cf.go_offline()
plt.style.use('classic')
%matplotlib inline
```

```python
df1 = pd.read_csv("data/29AUG.csv")
```

```python
df1.columns
```

```python
myP = ["#c26b00", "#242526", "#82096e", "#103009", "#34495e", "#c26b00", "#c2a200", "#0015b8", "#78282a", "#fc030b"]
```

```python
g = sns.catplot(x="Shifts_Quality", y="RTT_avg", col="Hosts_Competitions",palette=myP, data=df1, hue='Resolutions', height=5, aspect=.8)
g.set_axis_labels("Quality Shifts", "RTT Avg")
```

```python
g = sns.catplot(x="Shifts_Quality", y="RTT_stdev", col="Hosts_Competitions",palette=myP, data=df1,hue='Resolutions', height=5, aspect=.8)
g.set_axis_labels("Quality Shifts", "RTT Std")
```

```python
df1['RTT_stdev'] = df1['RTT_stdev'].astype(np.int64)
```

```python
g = sns.catplot(x="RTT_stdev", y="Avg_Bit_Rate", col="Hosts_Competitions", palette=myP,data=df1, height=5, aspect=.8,hue="Resolutions")
g.set_axis_labels("RTT_Std", "Average Bit Rate")
```

```python
g = sns.catplot(x="RTT_stdev", y="Total_Stall_Events",col="Hosts_Competitions", palette=myP,data=df1, hue="Resolutions", height=5, aspect=.8)
g.set_axis_labels("RTT_Std", "Total Stall Events")
```

```python
g = sns.catplot(x="RTT_stdev", y="Total_Stalls", hue="Resolutions",col="Hosts_Competitions",palette=myP,data=df1, height=5, aspect=.8)
g.set_axis_labels("RTT_Std", "Stall Duration in ms")
```

```python
g = sns.catplot(x="RTT_stdev", y="RTT_avg",col="Hosts_Competitions", hue="Resolutions",palette=myP, data=df1, height=5, aspect=.8)
g.set_axis_labels("RTT Avg", "RTT STD")
```
