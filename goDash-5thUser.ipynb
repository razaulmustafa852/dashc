```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("percentile - Sheet3.csv")
```

```python
df.head()
```

```python
data_5MB = df[(df['Bandwidth'] == 5)]
data_10MB = df[(df['Bandwidth'] == 10)]
data_15MB = df[(df['Bandwidth'] == 15)]
```

```python
bestCase = df[(df['Delay'] == 10)]
```

```python
worstCase = df[(df['Delay'] == 100)]
```

```python
g = sns.FacetGrid(bestCase, col="Bandwidth", hue="Competition",hue_kws={"marker": ["*", "o","s"]})
g.map(plt.scatter, "Percentile", "MOS", alpha=.7)
g.add_legend(title= 'Nodes');
g.set_axis_labels("Percentile", "Mean Opinion Score")
axes = g.axes.flatten()
axes[0].set_title("5 mbit/s")
axes[1].set_title("10 mbit/s")
axes[2].set_title("15 mbit/s")
plt.subplots_adjust(top=0.80)
g.fig.suptitle('Best Case: Delay 10ms & Node = 5') # can also get the figure from plt.gcf()
```

```python
g = sns.FacetGrid(worstCase, col="Bandwidth", hue="Competition",hue_kws={"marker": ["*", "o","s"]})
g.map(plt.scatter, "Percentile", "MOS", alpha=.7)
g.add_legend(title= 'Nodes');
g.set_axis_labels("Percentile", "Mean Opinion Score")
axes = g.axes.flatten()
axes[0].set_title("5 mbit/s")
axes[1].set_title("10 mbit/s")
axes[2].set_title("15 mbit/s")
plt.subplots_adjust(top=0.80)
g.fig.suptitle('Worst Case: Delay 100ms & Node = 5') # can also get the figure from plt.gcf()
```

```python

g= sns.barplot(x="Percentile", y="MOS", hue="Competition", data=data_5MB)
g.set_title('Case: 5 mbit/s, delay = 10ms, Nodes [5, 10, 15]')
g.set_ylabel('Mean Opinion Score')
g.set_xlabel('Hosts')
#L=plt.legend()
#L.get_texts().set_text('make it short')
g.legend(bbox_to_anchor=(1, 1.03))
```

```python
g= sns.barplot(x="Percentile", y="MOS", hue="Competition", data=data_10MB)
g.set_title('Case: 10 mbit/s, delay = 10ms, Nodes [5, 10, 15]')
g.set_ylabel('Mean Opinion Score')
g.set_xlabel('Hosts')
#L=plt.legend()
#L.get_texts().set_text('make it short')
g.legend(bbox_to_anchor=(1, 1.03))
```

```python
g= sns.barplot(x="Percentile", y="MOS", hue="Competition", data=data_15MB)
g.set_title('Case" 15 mbit/s, delay = 10ms, Nodes [5, 10, 15]')
g.set_ylabel('Mean Opinion Score')
g.set_xlabel('Hosts')
#L=plt.legend()
#L.get_texts().set_text('make it short')
g.legend(bbox_to_anchor=(1, 1.03))
```

```python
g = sns.barplot(x="Competition", y="MOS", hue="Percentile", data=data_5MB)
g.legend(bbox_to_anchor=(1, 1.03))
g.set_title('Case: 5 mbit/s, delay = 10ms')
g.set_ylabel('Mean Opinion Score')
g.set_xlabel('Hosts')
```

```python
g = sns.catplot(x="Percentile", y="MOS",
                 hue="Competition", col="Bandwidth",
                 data=bestCase, kind="bar",
                height=4, aspect=.9);
#g.add_legend(title= 'Nodes');
g.set_axis_labels("Chunks MOS percentile", "Mean Opinion Score")
axes = g.axes.flatten()
axes[0].set_title("5 mbit/s")
axes[1].set_title("10 mbit/s")
axes[2].set_title("15 mbit/s")
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Best Case: Delay 10ms & Node = 5') # can also get the figure from plt.gcf()
```
